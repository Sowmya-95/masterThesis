````python
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import os
import pandas as pd

# Path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = (r'/usr/bin/tesseract')

# Path to the input folder containing PDF files in sowmya_non_bail_docs folder
pdf_folder = '/content/drive/MyDrive/NLPMasterthesis/sowmya docs/sowmya_non_bail/non_bail_1' # Update this with your directory


# Output folder path for sowmya_non_bail_docs folder
output_folder_bail_docs = '/content/drive/MyDrive/NLPMasterthesis/Tesseract-OCR/Documents/output_folder_sowmya_non_bail_1' # Update this with your directory
os.makedirs(output_folder_bail_docs, exist_ok=True)

# Get a list of all PDF files in the folder
pdf_files = [file for file in os.listdir(pdf_folder) if file.lower().endswith('.pdf')]

# Create a DataFrame to store the document ID mapping
document_mapping = pd.DataFrame(columns=['Document_ID', 'File_Name'])

# Perform OCR on each PDF file
for i, pdf_file in enumerate(pdf_files):
    # Construct the full path to the PDF file
    pdf_path = os.path.join(pdf_folder, pdf_file)
    # Check if the document ID already exists in the mapping DataFrame
    document_id = f'DOC_{i + 1}'
    if document_id in document_mapping['Document_ID'].values:
        print(f"Skipping duplicate document ID - {document_id}")
        continue  # Skip processing this file

    # Convert PDF to images
    images = convert_from_path(pdf_path)

    # Perform OCR on each image
    text_content = ''
    for j, image in enumerate(images):
        # Save the image to a temporary file
        temp_image_path = f'temp_image_{i}_{j}.png'
        image.save(temp_image_path, 'PNG')

        # Perform OCR on the image
        image_text = pytesseract.image_to_string(Image.open(temp_image_path), lang='eng+hin',config='--psm 6 --oem 1')

        # Append the extracted text to the result
        text_content += image_text

        # Remove the temporary image file
        os.remove(temp_image_path)

    # Assign a document ID based on the index (i)
    document_id = f'DOC_{i + 1}'
    document_mapping = pd.concat([document_mapping, pd.DataFrame({'Document_ID': [document_id], 'File_Name': [pdf_file]})],
                                 ignore_index=True)


    # Create an output text file for each PDF
    output_file_path = os.path.join(output_folder_bail_docs, f'{document_id}.txt')
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(text_content)

    print(f"Text extracted from {pdf_file} and saved to {output_file_path}")

# Save the document mapping to an Excel file
excel_output_path = '/content/drive/MyDrive/NLPMasterthesis/Tesseract-OCR/Documents/document_mapping_sowmya_non_bail_1.xlsx' # Update this with your directory
document_mapping.to_excel(excel_output_path, index=False)
print(f"Document mapping saved to {excel_output_path}")
````
